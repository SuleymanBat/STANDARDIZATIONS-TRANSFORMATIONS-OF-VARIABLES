# We prepare three random lists as an numpay array
# and then we transform it into dataframe and float type

```python
V1 = np.array([1,3,5,6,8])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,7,8,33,4])
df= pd.DataFrame({'V1':V1,
                  'V2': V2,
                  'V3': V3})
df = df.astype(float)
df
```

# THE PROCESS OF TRANSFORMATIONS

# The process of transformation changes the values of variables permanently. But when you compare it with the standardization is not.

```python
import pandas as pd
df = pd.read_csv('../input/seaborn-tips-dataset/tips.csv')
df.head()
```

# LABEL ENCODING (0 - 1)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(df['sex'])
```

*Now we have transformed the categorical variables of `sex` column into range from 0 to 1 by using LabelEncoder function.*

```python
le.inverse_transform([0,1])
```

*if you have  forgotten what the first meaning of the numbers as a categorical you can look after by using inverse transform method like this. In this example you see 0 represants the females, 1 represants the males.*

```python
le.fit_transform(df['day'])
```
